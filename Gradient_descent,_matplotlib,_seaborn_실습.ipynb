```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D


def f(a,b):
	return (3/5)*a**4 + (2*b)**2 + np.cos(a*b)

def grad_f(a,b):
  h=1e-5
  return [(f(a+h,b) - f(a-h,b))/(2*h),(f(a,b+h) - f(a,b-h))/(2*h)]


def gradient_descent(theta0, iters, alpha):
	history = [theta0]
	theta = theta0
	for i in range(iters):
		gradient = grad_f(theta[0],theta[1])
		theta = [a - alpha*b for a,b in zip(theta, gradient)]
		history.append(theta)
	return history

history = gradient_descent(theta0 = [-3,-3], iters =40, alpha = 0.05)

fig = plt.figure(figsize=(10, 8))
ax = fig.gca(projection='3d')

a = np.arange(-3, 3, 0.25)
b = np.arange(-3, 3, 0.25)
a, b = np.meshgrid(a, b)
c = f(a,b)
surf = ax.plot_surface(a, b, c, rstride=1, cstride=1, alpha=0.3, 
		 linewidth=0, antialiased=False, cmap='rainbow')


a = np.array([x[0] for x in history])
b = np.array([x[1] for x in history])
c =  f(a,b)

ax.scatter(a, b, c, color="violet"); 

print(c)


plt.xlabel('Feature A')
plt.ylabel('Feature B')

plt.show()
```
